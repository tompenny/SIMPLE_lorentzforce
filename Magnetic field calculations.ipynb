Have now learnt we can use flat wire to make coils. This means we can fit way more in. Also want to consider putting coils outside of chamber. Let's start with just one big coil above the trap. Doing 20 turns is very reasonable here (that's what they have in Haystack) so let's start with that.

```python
u0 = 4*np.pi*10**(-7) # Magnetic permiability 
I = 10 # Current
R = np.linspace(4*10**(-2), 50*10**(-2)) # Range of coil radii to try 
z = 6*10**(-2)
n = 20
```

```python
B = 20*u0*I*R**2/2/(R**2+z**2)**(3/2)/10**(-4) # field from coil n0
```

```python
plt.plot(R*100, B, 'o')
plt.ylabel('Magnetic field (G)')
plt.xlabel('Coil radius (cm)')
```

```python
z = 6*10**(-2) # Distance from particle
ns = range(1, 200, 1)
R0l = np.array([5, 10, 15, 20, 25, 30, 35, 40, 45, 50])*10**(-2)
B3 = []
for R0 in R0l:
    B2 = [] # Magnetic field for each max coil number
    for n in ns:
        B_tmp2 = 0
        for n0 in range(n):
            B_tmp = u0*I*(R0+(n0)*0.5*10**(-3))**2/2/((R0+(n0)*0.5*10**(-3))**2+z**2)**(3/2)/10**(-4) # Magnetic field from coil n0
            B_tmp2 += B_tmp
        B2.append(B_tmp2)
        if n == 20:
            B3.append(B_tmp2)
    plt.plot(ns, B2, 'o', label = str(R0*100) +" cm")
plt.ylabel('B field (G)')
plt.xlabel('Number of coils')
plt.legend()
```

```python
# For a single loop how big/small till we get 1 ohm reactance at 20 Hz (this is what our power amplifier can do)?

u0 = 4*np.pi*10**(-7) # Magnetic permiability 
P = 10000
R = np.linspace(4*10**(-2), 25000*10**(0)) # Range of coil radii to try 
f = 20

L = (u0*np.pi*R/2*np.sqrt(P/f/2/np.pi))**(2/3)

I = np.sqrt(P/2/np.pi/f/L)
B = u0*I/2/R
```

```python
# Let's do multiple coils in a sprial
# Self inductance should cause exponential rise in total inductance

u0 = 4*np.pi*10**(-7) # Magnetic permiability 
I = 10 # Current
ns = range(1, 200, 1)
R0l = np.array([5, 10, 15, 20, 25, 30, 35, 40, 45, 50])*10**(-2)
L3 = []
for R0 in R0l: # scan through initial loop sizes
    L2 = [] # Total inductance of coil
    for n in ns:
        L_tmp = 0
        for n0 in range(n):
            for n1 in range(n):
                L_tmp2 = u0*I/2/(R0+(n0)*0.5*10**(-3))*np.pi*(R0+(n1)*0.5*10**(-3))**2
                L_tmp += L_tmp2
        L2.append(L_tmp)
    L3.append(L2)
```
